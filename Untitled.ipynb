# Laod data

In this part we read the mrcfile for emd 2472 and sample 1000 point from it. The sampling algorithm is very naive I just did it for the test. In the chart red dots are original map while blue dots are our samples.


```python
mymap = mrcfile.open('emd_2472.map')
n = mymap.data.shape[0]
m = mymap.data.shape[1]
l = mymap.data.shape[2]

x_gen = []
y_gen = []
z_gen = []
for i in range(n):
    for j in range(m):
        for k in range(l):
            if mymap.data[i,j,k] > 94.5:
                x_gen.append(i)
                y_gen.append(j)
                z_gen.append(k)

num = len(x_gen)
x =[]
y = []
z = []
# for i in range(1000):
#     index = random.randint(0, num - 1)
#     x.append(x_gen[index])
#     y.append(y_gen[index])
#     z.append(z_gen[index])
step_size = 2
for i in range(len(x_gen)):
    if x_gen[i] % step_size == 0 and y_gen[i] % step_size == 0 and z_gen[i] % step_size == 0:
        x.append(x_gen[i])
        y.append(y_gen[i])
        z.append(z_gen[i])

        
print("Sampled %d points out of %d points which had higher density than %0.1f"%(len(x),num,94.5))

fig = plt.figure()
ax = fig.gca(projection='3d', adjustable='box')

ax.scatter(x_gen, y_gen, z_gen,c=[(1,0,0,0.05)],  marker='o')
ax.scatter(x, y, z,  marker='o')
plt.show()
```

# Some rotation and translation

Just add a simple rotation and translation to it.

```python
theta = math.pi/7
t = [-1,2,1]
xr = []
yr = []
zr = []
for i in range(len(x)):
    xr.append(x[i]*math.cos(theta) - y[i]*math.sin(theta) + t[0])
    yr.append(x[i]*math.sin(theta) + y[i]*math.cos(theta) + t[1])
    zr.append(z[i] + t[2])
R = [[math.cos(theta),-math.sin(theta),0],[math.sin(theta), math.cos(theta),0],[0,0,1]]

print("Translation vector is:" + str(t))
print("Rotation matrix is:" + str(R))
```

# Optimal Transport

Apply OT to these two models. OT algorithm is emd just to ignore the complexity of a point be mapped to several point. I'll change it in the next expriment.

```python
def OT(x, y, z, xr, yr, zr):
    a = []
    b = []
    M = []
    for i in range(len(x)):
        M.append([])
        a.append(1)
        b.append(1)
        for j in range(len(x)):
            M[i].append((x[i] - xr[j])**2 + (y[i] - yr[j])**2 + (z[i] - zr[j])**2)

    #P = ot.sinkhorn(a, b, M, 10)
    P = ot.emd(a, b, M)
    print('optimal transport finished')
    return P

P = OT(x, y, z, xr, yr, zr)
c_tot = 0
c_cor = 0

for i in range(len(x)):
    for j in range(len(x)):
        if P[i][j] != 0:
            c_tot += 1
            if i == j:
                c_cor += 1

print('Out of %d correspondances %d are correct (%.1f%%)'%(c_tot,c_cor,c_cor/c_tot*100))
```

# Inferring a rigd body transfomation

Here you can see the rotation matrix and translation vector inferred from previouse part.

```python
d = []
m = []
dbar = np.zeros(3)
mbar = np.zeros(3)
for i in range(len(x)):
    for j in range(len(x)):
        if P[i][j] != 0:
            d.append(np.array([x[i],y[i],z[i]]))
            m.append(np.array([xr[j],yr[j],zr[j]]))
            dbar += 1/c_tot * np.array([x[i],y[i],z[i]])
            mbar += 1/c_tot * np.array([xr[j],yr[j],zr[j]])
            
            
dc = []
mc = []
for i in range(len(d)):
    dc.append(d[i]-dbar)
    mc.append(m[i]-mbar)
H = np.zeros((3,3))
for i in range(len(dc)):
    H += np.outer(mc[i],dc[i])
U, S, Vt = np.linalg.svd(H, full_matrices=True)
R = Vt.transpose().dot(U.transpose())
print("The rotation matrix is: \n" + str(R))
T = dbar - R.dot(mbar) 
print("The translation vector is: " + str(T))
```

# Illustraition

And in the end apply above matrices to data in order to get a nice visualization.

```python
fig = plt.figure()
ax = fig.gca(projection='3d', adjustable='box')
xres = []
yres = []
zres = []
for i in range(len(m)):
    res = R.dot(m[i]) + T
    xres.append(res[0])
    yres.append(res[1])
    zres.append(res[2])
    

ax.scatter(x, y, z,  marker='o')
ax.scatter(xres, yres, zres, marker='x')

plt.show()
```
