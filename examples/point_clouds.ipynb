## 2. Sampling
To sample point clouds from a specific map you could use `sample` function provided in `utils.py`.

```python
x, y, z = sample('Data/emd_1717.map', 72, 500, random_seed=1)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d', adjustable='box')
ax.scatter(x, y, z,  marker='o')
plt.show()
```
